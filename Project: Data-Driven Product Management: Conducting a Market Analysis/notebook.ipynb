You are a product manager for a fitness studio and are interested in understanding the current demand for digital fitness classes. You plan to conduct a market analysis in Python to gauge demand and identify potential areas for growth of digital products and services.

| Column     | Description              |

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
"""
# To check contents of current dir:
import os
print(os.listdir('.'))  # List files/folders in current directory
"""
# Get data
workout_df = pd.read_csv('data/workout.csv')
three_keywords_df = pd.read_csv('data/three_keywords.csv')
workout_geo_df = pd.read_csv('data/workout_geo.csv')
three_keywords_geo_df = pd.read_csv('data/three_keywords_geo.csv')
```

```python
# When was the global search for 'workout' at its peak?
max = 0
year_str = ""
for _, row in workout_df.iterrows():
    if row[1] > max:
        year_str = str(row[0])
        max = row[1]

year_str = year_str[:4]
```

```python
# What was the most popular keyword during covid and right now?
three_keywords_df

def get_max(dict):
    """
    Receives a dictionary of maxes
    
    Checks which is the biggest by sorting the dict in descending order
    and retrieving the first key-value pair

    Returns most popular key
    """
    return sorted(dict, key=dict.get, reverse=True)[0]
    
peak_covid_list = []
current_list = []

for row in three_keywords_df.itertuples():
    pop_keywords = {"home workout":row.home_workout_worldwide,
                    "gym workout":row.gym_workout_worldwide,
                    "home gym":row.home_gym_worldwide}
    month = str(row.month[5:])
    year = str(row.month[:4])
    
    if "2020-03" <= str(row.month) and str(row.month) <= "2020-12":
        peak_covid_list.append(get_max(dict=pop_keywords))
    else:
        current_list.append(get_max(dict=pop_keywords))

peak_covid = Counter(peak_covid_list).most_common(1)[0][0]
current = Counter(current_list).most_common(1)[0][0]
print(peak_covid)
print(current)
```

```python
# Among US, Australia and Japan, which country is the most interested in workouts?

top_country = ""
dict = {"United States":0,
       "Australia":0,
       "Japan":0}
for row in workout_geo_df.itertuples():
    if row.country == "United States":
        dict[row.country] = row.workout_2018_2023
    if row.country == "Australia":
        dict[row.country] = row.workout_2018_2023
    if row.country == "Japan":
        dict[row.country] = row.workout_2018_2023

    top_country = get_max(dict)
```

```python
# Philippines or Malaysia is the better client for the home workouts offering?
home_workout_geo = ""
dict = {"Philippines":0,
       "Malaysia":0}

for row in three_keywords_geo_df.itertuples():
    if row.Country == "Philippines":
        dict[row.Country] = row.home_workout_2018_2023
    if row.Country == "Malaysia":
        dict[row.Country] = row.home_workout_2018_2023

    home_workout_geo = get_max(dict)
```
